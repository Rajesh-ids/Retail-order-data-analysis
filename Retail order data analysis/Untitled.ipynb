```python
import pandas as pd
df = pd.read_csv("C:\\Users\\Lenovo\\project\\Retail order data analysis\\extracted_files\\orders.csv")
df
```

```python
#Handling Missing Values: Replace missing numerical values with defaults like 0 or drop rows with critical missing fields.
df.fillna(0, inplace = True)
```

```python
#Renaming Columns: Standardize column names for clarity and compatibility with SQL databases (e.g., converting Order ID to order_id).
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ","_")
```

```python
#Trimming Spaces: Remove trailing spaces from text fields.

text_col = df.select_dtypes(include = "object").columns
df[text_col] = df[text_col].apply(lambda x : x.astype(str).str.rstrip()) 
df
```

```python
#CALCULATING DISCOUNTS---derive new columns discount , sale price and profit
```

```python
df["discount"] = (df["list_price"] * df["discount_percent"])/100
df["sale_price"] = df["list_price"] - df["discount"]
df["profit"] = (df["sale_price"] - df["cost_price"]) * df["quantity"]
df["total_profit"] = df["profit"] * df["quantity"]

df.head()
```

```python
order_table = df.iloc[:, :9]
sales_table = df.iloc[:, 9:]
#order_table.head()
#sales_table.tail()
```

```python
sales_table.insert(0,"sales_id",range(101,101+len(sales_table)))
sales_table.insert(1,"order_id",order_table["order_id"])
```

```python
myconnection = pymysql.connect(host = '127.0.0.1', user = 'root', passwd= 'Admin@123')
```

```python
#create database
myconnection.cursor().execute("create database retail_order")
```

```python
#insert order table

df1 = order_table
database_name = "retail_order"
table_name = "order_table"
dtype_mapping = {
    'int64': 'INTEGER',
    'float64': 'FLOAT',
    'object': 'TEXT',
    'datetime64[ns]': 'DATETIME',
    'bool': 'BOOLEAN'
}
columns = ",".join(f"{col} {dtype_mapping[str(dtype)]}" for col, dtype in zip(df1.columns, df1.dtypes))

create_table_query = f"CREATE TABLE {database_name}.{table_name} ({columns});"
myconnection.cursor().execute(create_table_query)

for i in range(0,len(df1)):
    myconnection.cursor().execute(f"insert into {database_name}.{table_name} values {tuple(df1.iloc[i])}")
    myconnection.commit()
```
